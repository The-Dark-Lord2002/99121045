```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import cifar10
from minisom import MiniSom
```

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Normalize the data
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

# Flatten the images
x_train_flattened = x_train.reshape(x_train.shape[0], -1)
x_test_flattened = x_test.reshape(x_test.shape[0], -1)
```

```python
class SOM:
    def __init__(self, m, n, dim, num_iterations, alpha=None, sigma=None):
        self.m = m
        self.n = n
        self.dim = dim
        self.num_iterations = num_iterations
        self.alpha = alpha if alpha else 0.3
        self.sigma = sigma if sigma else max(m, n) / 2.0
        self.weights = np.random.rand(m, n, dim)
        self._init_time_decay()

    def _init_time_decay(self):
        self.time_constant = self.num_iterations / np.log(self.sigma)

    def _decay(self, initial_value, iteration):
        return initial_value * np.exp(-iteration / self.time_constant)

    def _neighborhood_function(self, distance, sigma):
        return np.exp(-distance**2 / (2 * sigma**2))

    def _calculate_distance(self, x, y):
        return np.linalg.norm(x - y, axis=-1)

    def _find_bmu(self, sample):
        distances = self._calculate_distance(self.weights, sample)
        bmu_index = np.unravel_index(np.argmin(distances), (self.m, self.n))
        return bmu_index

    def _update_weights(self, sample, bmu_index, iteration):
        learning_rate = self._decay(self.alpha, iteration)
        sigma = self._decay(self.sigma, iteration)
        for i in range(self.m):
            for j in range(self.n):
                distance = np.linalg.norm(np.array([i, j]) - np.array(bmu_index))
                neighborhood_value = self._neighborhood_function(distance, sigma)
                self.weights[i, j] += learning_rate * neighborhood_value * (sample - self.weights[i, j])

    def train(self, data):
        for iteration in range(self.num_iterations):
            sample = data[np.random.randint(0, data.shape[0])]
            bmu_index = self._find_bmu(sample)
            self._update_weights(sample, bmu_index, iteration)

    def map_vects(self, data):
        mapped = np.array([self._find_bmu(sample) for sample in data])
        return mapped
```

```python
# Define a function to plot the SOM
def plot_som(som, data, labels, title):
    plt.figure(figsize=(10, 10))
    plt.pcolor(som.distance_map().T, cmap='coolwarm')  # Distance map as background
    plt.colorbar()
    markers = ['o', 's', 'D', '^', 'v', '>', '<', 'p', 'h', '*']
    colors = ['C0', 'C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'C9']
    for cnt, xx in enumerate(data):
        w = som.winner(xx)
        plt.plot(w[0] + 0.5, w[1] + 0.5, markers[labels[cnt][0]], markerfacecolor='None',
                 markeredgecolor=colors[labels[cnt][0]], markersize=12, markeredgewidth=2)
    plt.title(title)
    plt.show()

plot_som(som, x_test_flattened, y_test, "SOM of CIFAR-10")
```
